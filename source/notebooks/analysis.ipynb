```python
import scipy as sp
import psycopg2
import pandas as pd
from matplotlib import pyplot as plt
```

```python
try:
    connection = psycopg2.connect(
        host="postgres",
        database="taxi_db",
        user="root",
        password="qwerty",
        port=5432
    )
    print("Успешно подключено к базе данных", flush=True)
    query = "SELECT * FROM taxi_rides"
    rides_df = pd.read_sql(query, connection)
    #print(rides_df)
except Exception as e:
    print(f"Error: {e}")
```

```python
print("Основная статистика по поездкам:")
print(rides_df[["distance_km", "price", "duration_min", "rating"]].describe())
```

```python
average_price = rides_df.groupby("vehicle_type")["price"].mean().sort_values(ascending=False)
print("Средняя цена по типам автомобилей:")
print(average_price)
```

```python
plt.figure(figsize=(10,6))
rides_df["vehicle_type"].value_counts().plot(kind="bar")
plt.title("Распределение поездок по типам автомобилей")
plt.xlabel("Тип автомобиля")
plt.ylabel("Количество поездок")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(rides_df["distance_km"], rides_df["price"])
plt.title("Зависимость цены от расстояния")
plt.xlabel("Расстояние (км)")
plt.ylabel("Цена (руб)")
plt.show()
print("Так как данные генерировались слу")
```

```python
plt.figure(figsize=(10, 6))
rides_df['rating'].hist(bins=20)
plt.title('Распределение рейтингов поездок')
plt.xlabel('Рейтинг')
plt.ylabel('Количество')
plt.show()
```
